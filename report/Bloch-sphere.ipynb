```python
import qutip as qt
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure(figsize=(5, 5))
ax = fig.add_subplot(azim=-40, elev=30, projection="3d")

b = qt.Bloch(axes=ax)
b.point_marker = ["o"]
b.zlabel=["$z$", ""]
b.plot_axes()
vec = [[1, 0, 0], [0, 1, 0], [0, 0, 1]]
b.add_vectors(vec)

th = np.linspace(0, 2 * np.pi, 20)
alphas = np.linspace(1, 0, 20)
xp = np.cos(th)
yp = np.sin(th)
zp = np.zeros(20)

pnts = zip(xp, yp, zp)
for p, a in zip(pnts, alphas):
    b.add_points(p, alpha=a, colors="b")

ax.text(0, 0, 0.5, fr"$|\psi \rangle$", fontsize=20)

b.save("blochspere.pdf", "pdf")
```
